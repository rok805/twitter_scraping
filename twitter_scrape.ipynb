```
from twitterscraper import query_tweets
from datetime import datetime
import datetime as dt
import pandas as pd
import time
df_merge = pd.DataFrame(columns=['text', 'timestamp'])
 

def get_twitt(x,y):
    begin_date = datetime.strptime(x, '%m/%d/%Y').date()
    end_date = datetime.strptime(y, '%m/%d/%Y').date()

    # limit = 10000
    lang = 'english'
    tweets = query_tweets("bitcoin", begindate=begin_date, enddate=end_date ,lang=lang) # limit=limit
    
    df = pd.DataFrame(t.__dict__ for t in tweets)
    df = df[['text', 'timestamp']]
    return df

def get_data_frame(i,j,x,y):
    sdate = '{0}/{1}/2019'.format(i,j)
    edate = '{0}/{1}/2019'.format(x,y)
    df = get_twitt(sdate, edate)
    return df
```

```

Start_time = time.time()
 

for j in range(1, 2):

    sdate = '9/{}/2019'.format(j)

    edate = '9/{}/2019'.format(j + 1)

    df = get_twitt(sdate, edate)

    df_merge = pd.concat([df_merge, df])

    df_merge.to_excel('/content/gdrive/My Drive/2019_09_01,{}.xlsx'.format(j))

    df_merge = pd.DataFrame(columns=['text', 'timestamp'])

C_time = time.time() - Start_time 

print(C_time)
```
